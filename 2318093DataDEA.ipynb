# **1. Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Aktifkan akses Google Drive
from google.colab import drive
drive.mount('/content/drive')
```

# **2. Load Data Kemiskinan**

```python
file_path = '/content/drive/MyDrive/Colab Notebooks/kemiskinan.csv'
df = pd.read_csv(file_path)

# Lihat 5 data teratas
df.head()
```

# **3. Informasi Umum Dataset**

```python
df.info()
```

# **4. Mengecek Data Kosong dan Statistik Umum**

```python
print("Jumlah data kosong tiap kolom:\n", df.isnull().sum())
df.describe()
```

# **5. Mengurutkan Provinsi Berdasarkan Tingkat Kemiskinan**

```python
top_kemiskinan = df.sort_values(by='Kemiskinan', ascending=False).reset_index(drop=True)

# Ambil 10 provinsi dengan tingkat kemiskinan tertinggi
top10 = top_kemiskinan.head(10)
top10
```

# **6. Visualisasi 10 Provinsi dengan Kemiskinan Tertinggi**

```python
plt.figure(figsize=(10,6))
sns.barplot(data=top10, x='Provinsi', y='Kemiskinan', palette='Reds_r')

plt.title('10 Provinsi dengan Tingkat Kemiskinan Tertinggi', fontsize=14)
plt.xlabel('Provinsi')
plt.ylabel('Tingkat Kemiskinan (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# **7. Visualisasi Perbandingan Semua Provinsi**

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df.sort_values(by='Kemiskinan', ascending=False),
            x='Provinsi', y='Kemiskinan', palette='coolwarm')

plt.title('Perbandingan Tingkat Kemiskinan Antar Provinsi', fontsize=14)
plt.xlabel('Provinsi')
plt.ylabel('Tingkat Kemiskinan (%)')
plt.xticks(rotation=75, ha='right')
plt.tight_layout()
plt.show()
```

# **8. Heatmap Korelasi Pearson Antar Fitur**

```python
# Pilih hanya kolom numerik
numerik_df = df.select_dtypes(include=['float64', 'int64'])

# Membuat heatmap korelasi Pearson
plt.figure(figsize=(10, 8))
ax = sns.heatmap(numerik_df.corr(), annot=True, fmt=".2f", cmap='coolwarm')

# Menambahkan judul
plt.title('Heatmap Korelasi Pearson Antar Faktor Kemiskinan', fontsize=14)

# Menghilangkan label sumbu x dan y
ax.set(xlabel="", ylabel="")

plt.show()
```
